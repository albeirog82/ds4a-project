```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
airbnb_2015 = pd.read_excel('barrios_airbnb-bogota-2015-2019.xlsx', sheet_name='2015', dtype={'Cod_Barrio': object})
airbnb_2016 = pd.read_excel('barrios_airbnb-bogota-2015-2019.xlsx', sheet_name='2016', dtype={'Cod_Barrio': object})
airbnb_2017 = pd.read_excel('barrios_airbnb-bogota-2015-2019.xlsx', sheet_name='2017', dtype={'Cod_Barrio': object})
airbnb_2018 = pd.read_excel('barrios_airbnb-bogota-2015-2019.xlsx', sheet_name='2018', dtype={'Cod_Barrio': object})
airbnb_2019 = pd.read_excel('barrios_airbnb-bogota-2015-2019.xlsx', sheet_name='2019', dtype={'Cod_Barrio': object})
```

```python
airbnb = pd.concat([airbnb_2015, airbnb_2016, airbnb_2017, airbnb_2018, airbnb_2019])
airbnb.shape
```

```python
airbnb.head()
```

```python
def get_unique_id(row):
    tipo = str(row['Tipo_AlojM'])
    max_h = str(row['Max_HuéspedesP'])
    ban = str(row['BañosP'])
    hab = str(row['HabitacionesP'])
    #return hash(row['Cod_Barrio'] + row['Tipo_AlojM']) + row['Tipo_AlquM'])
    hashme = str(row['Cod_Barrio']) + tipo + hab + ban + max_h  
    return hash(hashme)

airbnb['airbnb_id'] = airbnb.apply(get_unique_id, axis=1)

```
